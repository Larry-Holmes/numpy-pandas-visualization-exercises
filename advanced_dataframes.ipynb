```python
# Create a column named mileage_difference like you did in the DataFrames exercises; 
# this column should contain the difference between highway and city mileage for each car.

mpg_df['mileage_difference'] = mpg_df.hwy - mpg_df.cty
mpg_df
```

```python
# Create a column named average_mileage like you did in the DataFrames exercises; 
# this is the mean of the city and highway mileage.

mpg_df['avg_mileage'] = round(2/(1/mpg_df.cty+1/mpg_df.hwy), 2)
mpg_df
```

```python
# Create a new column on the mpg dataset named is_automatic that holds boolean values 
# denoting whether the car has an automatic transmission.

mpg_df['is_automatic'] = np.where(mpg_df.trans.str.startswith('auto') == True, 'yes', 'no')
mpg_df
```

```python
# Using the mpg dataset, find out which which manufacturer has the best miles per 
# gallon on average?

mpg_df.groupby('manufacturer').avg_mileage.mean().sort_values(ascending = False)
```

```python
# Do automatic or manual cars have better miles per gallon?

mpg_df.groupby('is_automatic')[['avg_mileage']].mean().round(1)
```

```python
# Use your get_db_url function to help you explore the data from the chipotle database.
schema = 'chipotle'

query4 = '''
SELECT
*
FROM orders;
'''
orders_df = pd.read_sql(query4, get_db_url)
orders_df.head()
```

```python
# What is the total price for each order?

orders_df['item_price'] = orders_df.item_price.str.replace('$','').astype(float)
# orders_df
orders_total = orders_df.groupby('order_id').item_price.sum()
```

```python
orders_total.sample(15)
```

```python
# What are the most popular 3 items?

pop_3 = orders_df.groupby('item_name').quantity.sum().sort_values(ascending=False).head(3)
pop_3
```

```python
# Which item has produced the most revenue?

orders_df.groupby('item_name').item_price.sum().nlargest(1)
```
