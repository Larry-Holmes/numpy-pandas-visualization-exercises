```python
import matplotlib.pyplot as plt
from random import randint
import numpy as np
import math
```

```python
# 1.

x = list(range(50))
y = [n**2 - n + 2 for n in x]

plt.plot(x,y)
plt.annotate('origin', xy=(0,0), xytext=(-3,5), arrowprops ={'facecolor' : 'red'})
plt.show()
```

```python
# 2.

y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

plt.plot(x,y1)
plt.show()
plt.plot(x,y2)
plt.show()
plt.plot(x,y3)
plt.show()
plt.plot(x,y4)
plt.show()
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.

plt.figure(figsize =(12, 6))

plt.subplot(221)
plt.plot(x,y1)
plt.subplot(222)
plt.plot(x,y2)
plt.subplot(223)
plt.plot(x,y3)
plt.subplot(224)
plt.plot(x,y4)
plt.show()
```

```python
# 4. Combine the figures you created in the last step into one figure where each of 
# the 4 equations has a different color for the points. Be sure to include a legend 
# and an appropriate title for the figure.

plt.figure(figsize =(12, 6))


plt.plot(x,y1,c='red')

plt.plot(x,y2,c='orange')

plt.plot(x,y3,c='cyan')

plt.plot(x,y4, c='darkviolet')

plt.legend(['y1','y2','y3','y4'])
plt.title('Unique Equations')
plt.show()
```
