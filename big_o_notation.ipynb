```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = list(range(1,50))

y = [0*n + 1 for n in x]
y1 = [math.log(n) for n in x]
y2 = [n for n in x]
y3 = [n * math.log(n) for n in x]
y4 = [n**2 for n in x]
y5 = [2**n for n in x]
y6 = [math.factorial(n) for n in x]
y7 = [n**n for n in x]

plt.figure(figsize=(14,6))


plt.plot(x,y, label ='O(1)')

plt.plot(x,y1, label ='O(log(n))')

plt.plot(x,y2, label ='O(n)')

plt.plot(x,y3, label ='O(n log n)')

plt.plot(x,y4, label ='O($n^{2}$)')

plt.plot(x,y5, label ='O($2^{n}$)')

plt.plot(x,y6, label ='O(n!)')

plt.plot(x,y7, label ='O($n^{n}$)')
plt.legend()
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')


plt.xlim(0,20)
plt.ylim(0,10)
plt.show()
```
