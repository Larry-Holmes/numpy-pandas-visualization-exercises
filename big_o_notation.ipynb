```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = list(range(1,50))

y = [0*n + 1 for n in x]
y1 = [math.log(n) for n in x]
y2 = [n for n in x]
y3 = [n * math.log(n) for n in x]
y4 = [n**2 for n in x]
y5 = [2**n for n in x]
y6 = [math.factorial(n) for n in x]
y7 = [n**n for n in x]

plt.figure(figsize=(14,6))

# plt.subplot(181)
plt.plot(x,y)
# plt.subplot(182)
plt.plot(x,y1)
# plt.subplot(183)
plt.plot(x,y2)
# plt.subplot(184)
plt.plot(x,y3)
# plt.subplot(185)
plt.plot(x,y4)
# plt.subplot(186)
plt.plot(x,y5)
# plt.subplot(187)
plt.plot(x,y6)
# plt.subplot(188)
plt.plot(x,y7)
plt.legend(['y','y1','y2','y3','y4','y5','y6','y7'])
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')


plt.xlim(0,20)
plt.ylim(0,10)
plt.show()
```
